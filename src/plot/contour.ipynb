```python
import cdms2 as cdms
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from pathlib import Path
home = str(Path.home())
import sys
sys.path.append(home+"/my_utils/src/")
from plot import contour
```

```python
ds = cdms.open("/global/homes/z/zhangtao/climate_causal/ENSO/data/ncar_ncep_ts_194801_202010.nc")
tas = ds['air'][24:-10-12*3,:,:]
lat = tas.getLatitude()[:]
lon = tas.getLongitude()[:]
```

# PlateCarree plot

```python
crange = np.arange(-35,35.01,1)
lat_rgns = [-90, 90]
lon_rgns = [0, 360]

fig, axes = plt.subplots(1,1,figsize=(11,8.5),subplot_kw={'projection':  ccrs.PlateCarree(180)},squeeze=False)

contour.plot_2d_contourf_PlateCarree(fig, axes[0][0], tas[0,:,:], lat, lon, lat_rgns, lon_rgns,
                                         "Correlation, TS->ENSO",'',crange,cmap="RdBu_r")

```

```python
fig, axes = plt.subplots(1,1,figsize=(11,8.5),subplot_kw={'projection': ccrs.Robinson(180)},squeeze=False)
contour.plot_2d_contourf(fig, axes[0][0], tas[0,:,:], lat, lon, 
                                         "Correlation, TS->ENSO",'k',crange,cmap="RdBu_r")
```

```python
fig, axes = plt.subplots(1,1,figsize=(11,8.5),subplot_kw={'projection': ccrs.SouthPolarStereo()},squeeze=False)
contour.plot_2d_contourf_PolarStereo(fig, axes[0][0], tas[0,:,:], lat, lon, 
                                         "Correlation, TS->ENSO",'k',crange,cmap="RdBu_r", polar="S")
```

```python
fig, axes = plt.subplots(1,1,figsize=(11,8.5),subplot_kw={'projection': ccrs.NorthPolarStereo()},squeeze=False)
contour.plot_2d_contourf_PolarStereo(fig, axes[0][0], tas[0,:,:], lat, lon, 
                                         "Correlation, TS->ENSO",'k',crange,cmap="RdBu_r", polar="N")
```
